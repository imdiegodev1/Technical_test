# Feature engineering

```python
##Get transaction_date and transaction_time
new = df["trans_date_trans_time"].str.split(" ", n = 2, expand = True)
df['trans_date'] = new[0]
df['trans_time'] = new[1]

df.drop(columns =["trans_date_trans_time"], inplace = True)

df.head()
```

```python
##Convert dates as str object to datetime object
#trans_date_trans_time
#dob

df['trans_date'] = pd.to_datetime(df['trans_date'], format='%Y/%m/%d')
df['dob'] = pd.to_datetime(df['dob'], format='%Y/%m/%d')

df.info()
```

```python
from datetime import datetime, date

def age(born):
    
    today = date.today()
    return today.year - born.year - ((today.month, 
                                      today.day) < (born.month, 
                                                    born.day))

##Get Age
df['age'] = df['dob'].apply(age)

df.head()
```

```python
#categorize age
labels = ["{0} - {1}".format(i, i + 9) for i in range(0, 100, 10)]
df["age_group"] = pd.cut(df['age'] , range(0, 105, 10), right=False, labels=labels)

df.head()
```

```python
##Get State name
df['state_name'] = df['state'].replace(abbrev_to_us_state)
```

```python
##Get complete name
df['name'] = df['first']+' '+df['last']
df.head()
```

```python
#Get month and year
df['month'] = pd.DatetimeIndex(df['trans_date']).month
df['year'] = pd.DatetimeIndex(df['trans_date']).year
df['day_of_week'] = df['trans_date'].dt.day_name()
```

```python
##Check data by gender and fraud
import matplotlib.pyplot as plt


df_gen = df[['gender', 'is_fraud']]
df_gen['count'] = 1
df_gen = df_gen.groupby(['gender', 'is_fraud'])['count'].sum().reset_index()

df_gen
```

```python
df_genM = df_gen[df_gen.gender == 'M']
df_genF = df_gen[df_gen.gender == 'F']
```

```python
N = 2
ind = np.arange(N)
width = 0.3
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(ind, df_genM['count'], width, color='r')
ax.bar(ind, df_genF['count'], width, bottom=df_genM['count'], color='b')
ax.set_ylabel('# Register')
ax.set_title('Fraud by Gender plot')
ax.set_xticks(ind, ('No_fraud', 'Fraud'))

ax.legend(labels=['Men', 'Women'])
plt.show()
```

With this first plot we can see the unbalanced data we have in the dataset. It is something common to find in data sets with this type of problems.

```python
##Evaluate age distribution
df_age = df[['age_group']]
df_age['count'] = 1

df_age = df_age.groupby(['age_group'])['count'].sum().reset_index()

df_age.head()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(df_age['age_group'], df_age['count'])
ax.set_ylabel('Freq')
ax.set_xlabel('Age Category')
ax.set_title('Age Category Distribution')

plt.show()
```

This plot let us know that we have a normal distribution with positive asymmetry. It is common to find people from 30 to 50 that get a service

```python
##Evaluate product distribution
df_prod = df[['category']]
df_prod['count'] = 1

df_prod = df_prod.groupby(['category'])['count'].sum().reset_index()
df_prod
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(df_prod['category'], df_prod['count'])
ax.set_ylabel('Freq')
ax.set_xlabel('Product Category')
ax.set_title('Product Category Distribution')

plt.xticks(rotation=45)
plt.show()
```
